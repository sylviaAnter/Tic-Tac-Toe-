```python
RED = "\033[91m" #-->X
YELLOW = "\033[93m" #-->O
RESET = "\033[0m" #-->Normal Colors
```

```python
def print_board(board):# board = list [1:9]
    for i in range(0, 9, 3):
        print(board[i], "|", board[i+1], "|", board[i+2])
```

#Coloring the Cell

```python
def color(cell):
    if cell == "X":
        return RED + "X" + RESET
    elif cell == "O":
        return YELLOW + "O" + RESET
    else:
        return cell
```

#Check the Winer

```python
def check_win(board,player):
  if board[0]==board[1]==board[2]==player:
    return True
  if board[3]==board[4]==board[5]==player:
    return True
  if board[6]==board[7]==board[8]==player:
    return True

  if board[0]==board[3]==board[6]==player:
    return True
  if board[1]==board[4]==board[7]==player:
    return True
  if board[2]==board[5]==board[8]==player:
    return True

  if board[2]==board[4]==board[6]==player:
    return True
  if board[0]==board[4]==board[8]==player:
    return True

  return False

```

```python

def game():
    board = [1,2,3,4,5,6,7,8,9]
    current_player = "X"

    for turn in range(9):
        colored_board = []
        for cell in board:
             colored_board.append(color(cell))

        #----------------------------------------------------
        print_board(colored_board)
        #----------------------------------------------------
        choice = int(input(f"Player {current_player}, choose position (1-9): "))
        #----------------------------------------------------
        while (choice < 1 or choice > 9 or board[choice-1] == "X" or board[choice-1] == "O"):
            choice = int(input("Invalid! choose again (1-9): "))
        #----------------------------------------------------
        board[choice-1] = current_player
        #----------------------------------------------------
        if check_win(board, current_player):
            colored_board = []
            for cell in board:
                 colored_board.append(color(cell))
            print_board(colored_board)
            print(f"Player {current_player} wins!")
            return

        #---------------------------------------------------
        # تبديل اللاعب
        if current_player == "X":
               current_player = "O"
        else:
              current_player = "X"
        #----------------------------------------------------

        colored_board = []
        for c in board:
              colored_board.append(color(c))
        #-------------------------------------------------------
    print_board(colored_board)
    print("play Again!")
```

```python
game()
```
